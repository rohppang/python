```python
# 유닛 6-3 서울 기온 데이터를 히스토그램으로 표현하기
import csv
```

```python
f = open('seoul.csv') #p29
f
```

```python
data = csv.reader(f) 
next(data)
```

```python
result = [] #비어있는 리스트를 준비
# 누락된 기온 ''
for row in data: #한 한줄 꺼낸다
    if row[-1] != '':  #누락되어 있지 않으면
        date = row[0]
        if date.split('-')[1] =='06':
            result.append(float(row[-1]))  #최고기온을 실수오 변환해서 저장
result
```

```python
plt.figure(figsize=(3,1.2))
n = plt.hist(result, bins=100, color='r')
```

```python
import pandas as pd
data = pd.read_csv('seoul.csv', encoding='cp949')
data = data.dropna() #전처리 : 누락된 갑이 있으면 그 행은 버린다
data.head() #상위 5건
```

```python
month = []
for i in range(12):
    month.append([])
x = np.array(data)
x.shape, x[0]
```

```python
sum(x[:, -1])
```

```python
np.mean(x[:, -1]), len(x)
```

```python
# 탐색적 프로그래밍
for i in range(len(x)):
    row = x[i]
    date = row[0]
    high = row[-1]
    mon = date.split('-')[1]
    # 1월은 0번째 위치에 저장해야 하므로 -1
    month[int(mon)-1].append(high)
    print(date, date.split('-') ,mon, high)
```

```python
plt.figure(figsize=(5,1.1))
m = plt.boxplot(month)
```
