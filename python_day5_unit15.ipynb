#268 pandas 로 인구 구조 분석하기

```python
import pandas as pd
df = pd.read_csv('age.csv', encoding='cp949', index_col=0)
df.shape
```

```python
df = df.dropna() #누락 행 제거
```

```python
df = df.replace(',','', regex=True)
```

```python
a = df.astype('int')
```

```python
# 인구를 촐 인구수로 나눈다. 인구수 > 비율
b = a.div(a.iloc[:,0], axis=0)  # 비율 데이터
c = b.iloc[:,2:]  #컬럼 2개 빼고
c.head(1)
```

```python
#2 궁금한 지역의 이름을 입력받는다.
name = input('궁금한 지역이 어디예요?')
name
```

```python
c.index.str.contains(name).sum()
mask = c.index.str.contains(name)
df2 = c[mask]
df2.head(2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family='Malgun Gothic')
df2.T.plot()
```

```python
df = c
x = df.sub(df2.iloc[0], axis=1) #원본과 내가 관심있는 지역의 0번째 행을 뺀다
y = np.power(x, 2) #제곱
z = y.sum(axis=1) #나이별 차이를 다 누적시킨다.
```

```python
i = z.sort_values().index[:5]
i
```

```python
df.loc[i].T.plot()
```
